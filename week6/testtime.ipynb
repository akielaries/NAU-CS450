```python
import time
from subprocess import call
```

```python
times = []
for _ in range(300):
    t = time.time()
    call('./testtime.out')
    times.append(time.time() - t)
```

```python
times[:10]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.hist(times)
plt.show()
```

```python
import numpy as np
import math
from scipy import stats
```

```python
times = np.array(times)
times.mean(), times.std()
```

```python
u_alpha = stats.norm.ppf(1 - 0.05)
u_alpha
```

```python
math.ceil(u_alpha**2 * times.var() / (0.05 * times.mean())**2)
```
